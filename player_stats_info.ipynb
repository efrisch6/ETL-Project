```python
import pandas as pd
from sqlalchemy import create_engine
import getpass
```

```python
skater = 'Resources/game_skater_stats.csv'
```

```python
player = 'Resources/player_info.csv'
```

```python
skater_df = pd.read_csv(skater, sep="|")
skater_df.head()
```

```python
skater_df = skater_df[['game_id','player_id','team_id','penaltyMinutes']]
```

```python
player_df = pd.read_csv(player, sep="|")
player_df.head()
```

```python
player_stats = skater_df.merge(player_df[['player_id','firstName','lastName','primaryPosition']], on='player_id', how='inner')
```

```python
player_stats
```

```python
player_stats.groupby(['game_id','team_id','primaryPosition',]).sum()
```

```python
p = getpass.getpass(prompt="Password: ")
rds_connection_string = f"postgres:{p}@localhost:5432/Hockey_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
skater_df.to_sql('skater_stats', con=engine, if_exists="append", index=False)
```
